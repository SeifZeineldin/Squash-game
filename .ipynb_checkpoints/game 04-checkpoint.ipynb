```python
import pygame
pygame.init()
#init display
screen_width=750
screen_hight=750
screen=pygame.display.set_mode((screen_width,screen_hight))
pg=pygame.image.load("game02.png")
font=pygame.font.SysFont("comicsans",25,True,True)
score_time=None
current_time=None

#////////////////////////////////////////////////////////////
class player():
     
    
    def __init__(self,color,x_player,y_player,step_x_player,step_y_player,hight,width,Is_my,score=0):
        self.x_player=x_player
        self.y_player=y_player
        self.step_x_player=step_x_player
        self.step_y_player=step_y_player
        self.hight=hight
        self.width=width
        self.color=color
        self.score=score
        self.Is_my=Is_my
    def stop(self,ball_x):
        if self.x_player>=(screen_width-self.width-20) or self.x_player<=0:
            return
        if ball_x>self.x_player:
            self.x_player-=15
            if self.x_player<0:
                self.x_player=0
                
        else:
            self.x_player+=15
    def set_x_y(self,x=15,y=screen_width//2+20):
        self.x_player=x
        self.y_player=y
    def get_Is_my(self):
        return self.Is_my
    def point(self):
        self.score+=1
    def get_score(self):
        return self.score
    def get_x_player(self):
        return self.x_player
    def get_y_player(self):
        return self.y_player
    def get_hight(self):
        return self.hight
    def get_width(self):
        return self.width
    def drow_player(self):
        pygame.draw.rect(screen,self.color,(self.x_player,self.y_player,self.width,self.hight))
    def move_player(self):
        keys=pygame.key.get_pressed()
        if keys[pygame.K_LEFT]:
            self.x_player -=self.step_x_player
            if self.x_player<0:
                self.x_player=0
        elif keys[pygame.K_RIGHT] :
            self.x_player +=self.step_x_player
            if self.x_player>(screen_width-self.width-20):
                self.x_player=screen_width-self.width-20
        elif keys[pygame.K_UP]:
            self.y_player-=self.step_y_player
            if self.y_player<20:
                self.y_player=20
        elif keys[pygame.K_DOWN]:
            self.y_player+=self.step_y_player
            if self.y_player>(screen_hight-self.hight-20):
                self.y_player=screen_hight-self.hight-20
    def AI_move(self,ball_x,ball_y,player_x,player_y):
            if self.x_player<0:
                self.x_player=0
            elif self.x_player>(screen_width-self.width-20):
                    self.x_player=screen_width-self.width-20
            elif self.y_player<20:
                    self.y_player=20
            elif self.y_player>(screen_hight-self.hight-20):
                    self.y_player=screen_hight-self.hight-20 
            #______________________________________________________
            if self.get_Is_my():
                if self.y_player < ball_y-(self.hight//2):
                    self.y_player +=self.step_y_player
                elif self.y_player > ball_y-(self.hight//2):
                    self.y_player -=self.step_y_player
                if ball_x-self.x_player>250:
                    self.x_player +=self.step_x_player
                elif self.x_player>ball_x:
                    self.x_player -=self.step_x_player




            else:
                if ball_x-self.x_player<50 and 0<=(screen_hight-self.hight-20)-self.y_player<=50 and self.x_player<=0 :
                    self.y_player -=self.step_y_player
                elif self.y_player-ball_y >100  :
                    self.y_player -=self.step_y_player
                elif  ball_y-self.y_player<100 :
                    self.y_player +=self.step_y_player
                #__________________________________________________

                if self.x_player<=0:
                    pass
                elif 0<ball_x-self.x_player<=50 and self.y_player<ball_y<self.y_player+self.hight:
                    self.x_player -=self.step_x_player
                elif 0<=self.x_player-ball_x<=50 and self.y_player<ball_y<self.y_player+self.hight:
                    self.x_player +=self.step_x_player

                
        

        

#////////////////////////////////////////////////////Ball class

class ball():
    def __init__(self):
        self.x_ball=screen_width//2
        self.y_ball=screen_hight//2
        self.x_step=15
        self.y_step=15
        self.reduis=10
    def set_x_y(self,x=screen_width//2,y=screen_hight//2):
        self.x_ball=x
        self.y_ball=y
    def drow_ball(self,screen):
        pygame.draw.circle(screen,(250, 240, 215), (self.x_ball,self.y_ball),self.reduis)
    def get_ball_x(self):
        return self.x_ball
    def get_ball_y(self):
        return self.y_ball
    def get_reduis(self):
        return self.reduis
    def move_ball(self):
        self.x_ball +=self.x_step
        self.y_ball+=self.y_step
        if self.x_ball>=(screen_width-20):
            self.x_step*=-1
        if self.y_ball<=20 or self.y_ball>=(screen_hight-20):
            self.y_step*=-1
#///////////////////////////////////////////////////Game
class game():
    ball=ball()
    player_1=player((235, 83, 83),60,300,15,15,85,15,True)
    AI_player=player((24, 116, 152),10,300,15,15,85,15,False)
    def game_start(self):
        global current_time , score_time
        self.ball.set_x_y()
        if self.player_1.get_Is_my():
            screen.blit(font.render("Your turn",True,(0,0,0)),(screen_width/2 - 55, screen_hight/2 - 20))
            self.player_1.set_x_y(50)
            self.AI_player.set_x_y()
        else :
            screen.blit(font.render("AI turn",True,(0,0,0)),(screen_width/2 - 55, screen_hight/2 - 20))
            self.player_1.set_x_y()
            self.AI_player.set_x_y(50)

        current_time=pygame.time.get_ticks()
        if current_time - score_time < 700:
            number_three = font.render("3",False,(0,0,0))
            screen.blit(number_three,(screen_width/2 - 10, screen_hight/2 + 20))
        elif 700 < current_time - score_time < 1400:
            number_two = font.render("2",False,(0,0,0))
            screen.blit(number_two,(screen_width/2 - 10, screen_hight/2 + 20))
        elif 1400 < current_time - score_time < 2100:
            number_one = font.render("1",False,(0,0,0))
            screen.blit(number_one,(screen_width/2 - 10, screen_hight/2 + 20))


        if current_time - score_time > 2100:
            self.ball.x_step*=-1
            score_time=None




    def game_mange(self):

        global score_time
        #////////

        #///////////////////////////////////////////
        if((self.ball.x_step==-15 and
            self.player_1.get_x_player()+(self.player_1.get_width()//2)<=self.ball.get_ball_x() <=self.player_1.get_x_player()+self.player_1.get_width())
           and(self.player_1.get_y_player()<=self.ball.get_ball_y()<=self.player_1.get_y_player()+self.player_1.get_hight())):
            if self.player_1.get_Is_my():
                self.player_1.Is_my=False
                self.AI_player.Is_my=True
                self.ball.x_step*=-1
                self.player_1.stop(self.ball.get_ball_x())
            else: 
                self.AI_player.point()
                self.player_1.Is_my=False
                self.AI_player.Is_my=True
                score_time=pygame.time.get_ticks()
        elif((self.ball.x_step==-15 and
            self.player_1.get_x_player()<=self.ball.get_ball_x() <=self.player_1.get_x_player()+self.player_1.get_width()//2)
           and(self.player_1.get_y_player()<=self.ball.get_ball_y()<=self.player_1.get_y_player()+self.player_1.get_hight())):
            if self.player_1.get_Is_my():
                self.player_1.Is_my=False
                self.AI_player.Is_my=True
                self.ball.y_step*=-1
                self.player_1.stop(self.ball.get_ball_x())
            else: 
                self.AI_player.point()
                self.player_1.Is_my=False
                self.AI_player.Is_my=True
                score_time=pygame.time.get_ticks()
        elif((self.ball.x_step==15 and
            self.player_1.get_x_player()<=self.ball.get_ball_x() <=self.player_1.get_x_player()+self.player_1.get_width()//2)
           and(self.player_1.get_y_player()<=self.ball.get_ball_y()<=self.player_1.get_y_player()+self.player_1.get_hight())):
            if self.player_1.get_Is_my():
                self.player_1.Is_my=False
                self.AI_player.Is_my=True
                self.ball.x_step*=-1
                self.player_1.stop(self.ball.get_ball_x())
            else: 
                self.AI_player.point()
                self.player_1.Is_my=False
                self.AI_player.Is_my=True
                score_time=pygame.time.get_ticks()
        elif((self.ball.x_step==15 and
            self.player_1.get_x_player()+(self.player_1.get_width()//2)<=self.ball.get_ball_x() <=self.player_1.get_x_player()+self.player_1.get_width())
           and(self.player_1.get_y_player()<=self.ball.get_ball_y()<=self.player_1.get_y_player()+self.player_1.get_hight())):
            if self.player_1.get_Is_my():
                self.player_1.Is_my=False
                self.AI_player.Is_my=True
                self.ball.y_step*=-1
                self.player_1.stop(self.ball.get_ball_x())
            else: 
                self.AI_player.point()
                self.player_1.Is_my=False
                self.AI_player.Is_my=True
                score_time=pygame.time.get_ticks()
        #//////////////////////////////////////////////////////////////
                
        elif ((self.ball.x_step==-15 and
            self.AI_player.get_x_player()+(self.AI_player.get_width()//2)<=self.ball.get_ball_x() <=self.AI_player.get_x_player()+self.AI_player.get_width())
              and(self.AI_player.get_y_player()<=self.ball.get_ball_y()<=self.AI_player.get_y_player()+self.AI_player.get_hight())):
            if self.AI_player.get_Is_my():
                self.player_1.Is_my=True
                self.AI_player.Is_my=False
                self.ball.x_step*=-1
                
            else:
                self.player_1.point()
                self.player_1.Is_my=True
                self.AI_player.Is_my=False
                self.AI_player.set_x_y()
                self.player_1.set_x_y()
                self.ball.set_x_y()
                score_time=pygame.time.get_ticks()
        elif ((self.ball.x_step==-15 and
            self.AI_player.get_x_player()<=self.ball.get_ball_x() <=self.AI_player.get_x_player()+self.AI_player.get_width()//2)
              and(self.AI_player.get_y_player()<=self.ball.get_ball_y()<=self.AI_player.get_y_player()+self.AI_player.get_hight())):
            if self.AI_player.get_Is_my():
                self.player_1.Is_my=True
                self.AI_player.Is_my=False
                self.ball.y_step*=-1
                
            else:
                self.player_1.point()
                self.player_1.Is_my=True
                self.AI_player.Is_my=False
                self.AI_player.set_x_y()
                self.player_1.set_x_y()
                self.ball.set_x_y()
                score_time=pygame.time.get_ticks()
                
                
        elif ((self.ball.x_step==15 and
            self.AI_player.get_x_player()+(self.AI_player.get_width()//2)<=self.ball.get_ball_x() <=self.AI_player.get_x_player()+self.AI_player.get_width())
              and(self.AI_player.get_y_player()<=self.ball.get_ball_y()<=self.AI_player.get_y_player()+self.AI_player.get_hight())):
            if self.AI_player.get_Is_my():
                self.player_1.Is_my=True
                self.AI_player.Is_my=False
                self.ball.y_step*=-1
                
            else:
                self.player_1.point()
                self.player_1.Is_my=True
                self.AI_player.Is_my=False
                self.AI_player.set_x_y()
                self.player_1.set_x_y()
                self.ball.set_x_y()
                score_time=pygame.time.get_ticks()
        elif ((self.ball.x_step==15 and
            self.AI_player.get_x_player()<=self.ball.get_ball_x() <=self.AI_player.get_x_player()+self.AI_player.get_width()//2)
              and(self.AI_player.get_y_player()<=self.ball.get_ball_y()<=self.AI_player.get_y_player()+self.AI_player.get_hight())):
            if self.AI_player.get_Is_my():
                self.player_1.Is_my=True
                self.AI_player.Is_my=False
                self.ball.x_step*=-1
                
            else:
                self.player_1.point()
                self.player_1.Is_my=True
                self.AI_player.Is_my=False
                self.AI_player.set_x_y()
                self.player_1.set_x_y()
                self.ball.set_x_y()
                score_time=pygame.time.get_ticks()
                
        elif self.ball.get_ball_x()<0 and self.player_1.get_Is_my():
            self.AI_player.point()
            self.player_1.Is_my=False
            self.AI_player.Is_my=True
            self.AI_player.set_x_y()
            self.player_1.set_x_y()
            self.ball.set_x_y()
            score_time=pygame.time.get_ticks()
            
        elif self.ball.get_ball_x()<0 and self.AI_player.get_Is_my():
            self.player_1.point()
            self.player_1.Is_my=True
            self.AI_player.Is_my=False
            self.AI_player.set_x_y()
            self.player_1.set_x_y()
            self.ball.set_x_y()
            score_time=pygame.time.get_ticks()
                
        elif self.ball.get_ball_x()<0 and self.player_1.get_Is_my():
            self.AI_player.point()
            self.player_1.Is_my=False
            self.AI_player.Is_my=True
            self.AI_player.set_x_y()
            self.player_1.set_x_y()
            self.ball.set_x_y()
            score_time=pygame.time.get_ticks()
            
        elif self.ball.get_ball_x()<0 and self.AI_player.get_Is_my():
            self.player_1.point()
            self.player_1.Is_my=True
            self.AI_player.Is_my=False
            self.AI_player.set_x_y()
            self.player_1.set_x_y()
            self.ball.set_x_y()
            score_time=pygame.time.get_ticks()
            
        
#////////////////////////////////Main
game=game()
run=True

while run:
    if game.player_1.get_score()<5 and game.AI_player.get_score()<5:
        pygame.time.delay(50)
        for event in pygame.event.get():
                   if event.type == pygame.QUIT:
                    run = False
        screen.blit(pg,(0,0)) 
        screen.blit(font.render("PLAYER :"+str(game.player_1.get_score())+"  AI:"+str(game.AI_player.get_score()),True,(0,0,0)),(screen_width//2-130,10))
        #ball Functions
        game.ball.drow_ball(screen)    
        game.ball.move_ball()
        #palyer Function
        game.player_1.drow_player()
        game.player_1.move_player()
        game.AI_player.drow_player()
        game.AI_player.AI_move(game.ball.get_ball_x(),game.ball.get_ball_y(),game.player_1.get_x_player(),game.player_1.get_y_player())
        #check pand
        game.game_mange()
        if score_time:
            game.game_start()


        # Drow lines borders
        pygame.draw.line(screen, (54, 174, 124),
                (0,0), (screen_width,0),20)
        pygame.draw.line(screen, (54, 174, 124),
                 (screen_width,0), (screen_width,screen_hight),20)
        pygame.draw.line(screen, (54, 174, 124),
                (0,screen_hight), (screen_width,screen_hight),20)
        pygame.display.update()
    else:
        if game.AI_player.get_score()>5:
            screen.blit(font.render("Yoy Win",True,(0,0,0)),(screen_width/2 - 55, screen_hight/2 - 20))

            break
        else:
            screen.blit(font.render("You loss",True,(0,0,0)),(screen_width/2 - 55, screen_hight/2 - 20))

            break

pygame.quit()
```
